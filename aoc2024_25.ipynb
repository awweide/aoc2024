```python
input_test = """#####
.####
.####
.####
.#.#.
.#...
.....

#####
##.##
.#.##
...##
...#.
...#.
.....

.....
#....
#....
#...#
#.#.#
#.###
#####

.....
.....
#.#..
###..
###.#
###.#
#####

.....
.....
.....
#....
#.#..
#.#.#
#####"""

output_test = 3
```

```python
import numpy as np
import itertools
```

```python
def parse(inp):
    keys = []
    locks = []
    for block in inp.split('\n\n'):
        list_block = keys if block.startswith('#####') else locks
        npa = np.array([list(row) for row in block.splitlines()])
        npa = npa == '#'
        list_block.append(npa)
    return keys,locks

def solve(inp):
    keys, locks = parse(inp)
    return sum([not np.any(np.logical_and(key,lock)) for (key,lock) in itertools.product(keys, locks) ])
```

```python
assert solve(input_test) == output_test
solve(input_full)
```
