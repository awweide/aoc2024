```python
import numpy as np
def parse_input(inp):
    return np.array([np.array(list(row)) for row in inp.split('\n')])

def get_all_rows_cols_diagonals(arr):
    l = [_ for _ in arr] #rows
    l += [_ for _ in np.transpose(arr)] #columns
    l += [np.diagonal(arr, i) for i in range(-arr.shape[1], arr.shape[0])] #normal diagonals
    l += [np.diagonal(np.fliplr(arr), i) for i in range(-arr.shape[1], arr.shape[0])] #reverse diagonals
    return l

def count_substring_in_string_as_list(lst, substring):
    return (''.join(lst)).count(substring) + (''.join(lst[::-1]).count(substring))

def solve_p1(inp, substring='XMAS'):
    return sum([count_substring_in_string_as_list(lst, substring) for lst in get_all_rows_cols_diagonals(parse_input(inp))])
```

```python
def validate_3x3(g, s):
    pos_diag = ''.join(np.diagonal(g, 0))
    neg_diag = ''.join(np.diagonal(np.fliplr(g), 0))
    #horizontal = ''.join(g[1,:])
    #vertical = ''.join(g[:,1])
    matches = [s, s[::-1]]

    if pos_diag in matches and neg_diag in matches: return True
    #if vertical in matches and horizontal in matches: rval = True
    return False

def solve_p2(inp, substring='MAS'):
    g = parse_input(inp)
    range_tuple = (g.shape[0] - 2, g.shape[1] - 2)
    return sum([validate_3x3(g[x:x+3,y:y+3], substring) for (x,y) in np.ndindex(range_tuple) ])
```
