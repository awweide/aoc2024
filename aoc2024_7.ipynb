```python
def parse_line(l):
    target, args = l.split(': ')
    args = [int(a) for a in args.split(' ')]
    return (int(target), args)
def parse_input(inp):
    return [parse_line(line) for line in inp.split('\n')]
```

```python
from itertools import product

#Brute force solution is slightly slow (30s?) for p2
def eval_line_bf(line, p2 = False):
    ops_from = ["+", "*"]
    if p2: ops_from.append("&")
    target, args = line
    for ops in product(ops_from, repeat = len(args) - 1):
        val = args[0]
        for op,arg in zip(ops, args[1:]):
            if op == "+": val += arg
            elif op == "*": val *= arg
            elif op == "&": val = int(str(val) + str(arg))
            else: raise ValueError
            if val > target: break #val can only grow during loop, early stopping saves some time
        if val == target: return target
    return False

def solve_bf(inp, p2=False):
    return sum([eval_line_bf(line, p2) for line in parse_input(inp)])
        
```
