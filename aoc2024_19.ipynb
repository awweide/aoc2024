```python
from collections import defaultdict
```

```python
def parse_input(inp):
    parts, orders = inp.split('\n\n')
    parts = parts.split(', ')
    orders = orders.split('\n')
    return parts, orders

#Deprecated: works, but it is way too slow
def solve_order_recursive(order, parts, part_max_length):
    if not order: return True    
    for i in range(1, max(part_max_length, len(order))):
        if order[:i] in parts:
            if solve_order_recursive(order[i:], parts, part_max_length): return True
    return False

def solve_order(order, parts, p2 = False):
    l_order = len(order)
    dicts_by_len_of_tails = []
    for i in range(0, l_order+1): dicts_by_len_of_tails.append(defaultdict(int))
    dicts_by_len_of_tails[l_order][order] = 1
    
    for l_tail in range(l_order, -1, -1):
        for tail,num in dicts_by_len_of_tails[l_tail].items():
            for part in parts:
                l_part = len(part)
                if tail[:l_part] == part:
                    dicts_by_len_of_tails[l_tail-l_part][tail[l_part:] if l_part < l_tail else " "] += num
    #We would use "" as key for the 0-length dict, but "" is not allowed as a key so we replace it with " "
    return dicts_by_len_of_tails[0][" "] if p2 else dicts_by_len_of_tails[0][" "] > 0

def solve_input(inp, p2=False):
    parts, orders = parse_input(inp)
    return sum([solve_order(order, parts, p2) for order in orders])
```
