```python
def parse_input(inp):
    n = defaultdict(set)
    for line in inp.split('\n'):
        f,t = line.split('-')
        n[f].add(t)
        n[t].add(f)
    return n

def get_largest_fcsn(network):
    fcsn_nodes = []
    fcsn_size = 0
    while network:
        #pick a node; assume this node is in fcsn; popping prunes the node right away
        node, node_connections = network.popitem()

        #If node has enough connections to possibly create a largest fcsn
        if len(node_connections) + 1 > fcsn_size:
            #create a subnetwork of connections with only nodes fully connected to node
            subnetwork = {cnode: network[cnode] & node_connections for cnode in node_connections}
            #find largest fcsn in this subnetwork
            sub_fcsn_nodes, sub_fcsn_size = get_largest_fcsn(subnetwork)
            #remember it if it's best candidate; add 1 because "node" is also in this subnetwork
            if sub_fcsn_size + 1 > fcsn_size: fcsn_nodes, fcsn_size = sub_fcsn_nodes + [node], sub_fcsn_size + 1
        
        #prune picked node from other connections
        for node_connected in node_connections: network[node_connected].remove(node)
        
    return fcsn_nodes, fcsn_size

def groups_from_connections(c):
    groups = set()
    for f, list_t in c.items():
        for t in list_t:
            for t2 in c[t]:
                if f in c[t2]:
                    groups.add(tuple(sorted([f,t,t2])))
    return groups

def solve_p1(inp):
    groups = groups_from_connections(parse_input(inp))
    return sum([any([sg.startswith('t') for sg in g]) for g in groups])

def solve(inp):
    network = parse_input(inp)
    fcsn_nodes, fcsn_size = get_largest_fcsn(network)
    fcsn_nodes = sorted(fcsn_nodes)
    password = ""
    for node in fcsn_nodes: password += "," + node
    return password[1:]
```
