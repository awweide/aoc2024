```python
def parse_line(line): return [np.array(list(map(int, i[2:].split(',')))) for i in line.split(' ')]
def parse_input(inp): return [parse_line(line) for line in inp.split('\n')]

def pos_at_time(p, v, t, size): return np.mod(p + v * t, size)
def quadrant_from_pos(p,size):
    if p[0] == size[0] // 2 or p[1] == size[1] // 2: return 0
    q = 1
    if p[0] > size[0] // 2: q += 1
    if p[1] > size[1] // 2: q += 2    
    return q

def period(p,v,size):
    for t in range(1,10**6):
        if np.all(pos_at_time(p,v,t,size) == p):
            return t 
    return None

def solve(inp, size, steps = 100):
    pv_list = parse_input(inp)
    quadrant_count = [0] * 5
    for p,v in pv_list:
        quadrant_count[quadrant_from_pos(pos_at_time(p,v,steps,size), size)] += 1
    return reduce(lambda x,y: x*y, quadrant_count[1:])

def solve_p2(inp, size):
    #Following finds solution i == 6620 by visual inspection of the steps with the  "fullest" lines
    g = np.array([[0] * size_full[0] for i in range(size_full[1])])
    pv_list = parse_input(input_full)
    mx = 0
    for i in range(0, 10500):
        mx = max(mx, grid_print(g, pv_list, size_full, i, mx))
    
    #Following reveals periodicity is 101*103 for all vectors, constraining search
    #for p,v in parse_input(inp):
    #    print(p, v, period(p,v,size))

def grid_print(g, pv_list, size, steps, mx = 0):
    g = g.copy()

    for p,v in pv_list:
        x,y = pos_at_time(p,v,steps,size)
        g[y,x] = 1
    
    MX = max(np.max(np.sum(g == 1, axis=1)), np.max(np.sum(g == 1, axis=0)))
    if MX > mx:
        print(MX, steps)
        for line in g: print("".join(map(lambda x: "X" if x == 1 else " ", line)))
    return MX
```
